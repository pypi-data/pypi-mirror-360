## This notebook compares Cine-Gaussian signals, taking into account also ellipticity, generated with Burst Waveform and cWB MDC

it gives back:
\begin{equation}
    h_{plus} = \frac{1}{\sqrt{2}} \frac{(1+\cos^2{(\iota)})}{2} \times CosineGaussian

\end{equation}

\begin{equation}
    h_{cross} = \frac{1}{\sqrt{2}} \cos{(\iota)} \times SineGaussian
    \nonumber
\end{equation}

### Packages

```python
import sys
import os
import numpy as np
sys.path.append(os.path.abspath('../'))
from burst_waveform.models.sine_gaussian import SineGaussianQEllipticity
from matplotlib import pyplot as plt

from gwpy.timeseries import TimeSeries
```

```python
frequency = 100 #np.random.uniform(10, 2000)
Q = 10 #np.random.uniform(1,100)              # duration = Q / (2*pi*frequency)

# Generate the burst with python burst waveform module
params = {
    "amplitude": 1.0,
    "frequency": frequency,
    "Q": Q,
    "ellipticity": 0.5,
}
model = SineGaussianQEllipticity(params)
h_plus, h_cross = model()

# plot the waveform
plt.figure(figsize=(10, 5))
plt.plot(h_plus.times, h_plus.value, label='h_plus')
plt.plot(h_cross.times, h_cross.value, label='h_cross')
plt.title('Sine Gaussian Q Ellipticity Waveform')
plt.xlabel('time [s]')
plt.ylabel('strain')
plt.legend()
plt.show()
```
