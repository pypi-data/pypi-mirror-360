```python
import groupedpaneldatamodels as gpdm
import numpy as np
```

```python
import pickle

example_data = pickle.load(open("groupedpaneldatamodels/dgp3-example-1749134059.pkl", "rb"))
example_data
```

```python
x_example = example_data[0]
y_example = np.reshape(example_data[1], (100, 50, 1))
f_example = example_data[2]
g_example = example_data[3]
```

```python
t = gpdm.GroupedInteractiveFixedEffects(
    y_example, x_example, 3, model="su_ju", use_bootstrap=True
).fit(n_boot=5)
display(t.params)
display(t.params_standard_errors)
display(t.get_confidence_intervals(0.99))
```
